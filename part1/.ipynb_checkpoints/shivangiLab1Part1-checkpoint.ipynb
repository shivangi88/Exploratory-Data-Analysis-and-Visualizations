# Problem 1 Analysis

## We can see from the graphs, we are getting different values for the sales2 vector. This is because we are creating the sales2 vector using rpois which creates random values from a poisson distribution.


```r

## Problem 1 ##

sales1<-c(12,14,16,29,30,45,19,20,16, 19, 34, 20)
sales2<-rpois(12,34)  # random numbers, Poisson distribution, mean at 34, 12 numbers
par(bg="cornsilk")

plot(sales1, col="blue", type="o", ylim=c(0,100), xlab="Month", ylab="Sales" )
title(main="Sales by Month")

lines(sales2, type="o", pch=22, lty=2, col="red")
grid(nx=NA, ny=NULL)
legend("topright", inset=.05, c("Sales1","Sales2"), fill=c("blue","red"), horiz=TRUE)
png('Problem_1_Monthly_Sales')
```


# Problem 2 Analysis

## From the sales data, we can assume that it is a month wise performance comparison of 2 sales. We have used the rainbow(12) keyword, It makes it visually easier to do a monthwise comparison. As we can see from the histogram, the sales2 have higher sales than sales1 for the month of July and in general sales2 has a better performance than sales1.



```r

## Problem 2 ##
sales<-read.table(file.choose(), header=T)
sales  # to verify that data has been read
barplot(as.matrix(sales), main="Sales Data", xlab="Months",ylab= "Total",beside=T, col=rainbow(12),ylim=c(0,100))
```


# Problem 3 Analysis

## From the boxplot, we can see that the majority of data is on the lower side of the range, so the data is skewed and may not be normally distributed. The median for the sales1 tends more towards the first quartile, this indicates that the data for sales1 is more concentrated near the first quartile. Whereas for sales2 the median is fairly near the center of the inter quartile range.


```r

## Problem 3

fn<-boxplot(sales,col=c("orange","green"),ylim=c(0,100))$stats
text(1.45, fn[3,2], paste("Median =", fn[3,2]), adj=0, cex=.7)
text(0.45, fn[3,1],paste("Median =", fn[3,1]), adj=0, cex=.7)
grid(nx=NA, ny=NULL)
```
